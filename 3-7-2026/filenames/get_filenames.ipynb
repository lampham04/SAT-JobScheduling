```python
# READ filename
import pandas as pd
from pathlib import Path
```

```python
def get_filenames_with_prec(folder_path):
    folder = Path(folder_path)

    filtered = [
        f.name
        for f in folder.glob("*.GSP")
        if f.stem.split("_")[-2] != "00"
    ]
    
    return filtered

def write_filenames_to_file(instance_names, instance_name_file_path):
    with open(instance_name_file_path, "w") as f:
        for name in instance_names:
            f.write(name + "\n")
```

```python
INSTANCE_FOLDER_PATH = "../datasets/30-S"
INSTANCE_NAME_FILE_PATH = "./30-S.txt"

instance_names = get_filenames_with_prec(INSTANCE_FOLDER_PATH)
len(instance_names)
write_filenames_to_file(instance_names, INSTANCE_NAME_FILE_PATH)
```
