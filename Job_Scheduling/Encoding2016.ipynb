```python
predecessors = {i: [] for i in range(1, n+1)}
for i in range(1, n+1):
    for s in successors[i]:
        predecessors[s].append(i)

predecessors
```

```python
from collections import deque

indeg = {i: len(predecessors[i]) for i in range(1, n+1)}
q = deque([i for i in range(1, n+1) if indeg[i] == 0])
topo = []

while q:
    u = q.popleft()
    topo.append(u)
    for v in successors[u]:
        indeg[v] -= 1
        if indeg[v] == 0:
            q.append(v)

if len(topo) != n:
    raise ValueError("Precedence graph has a cycle!")

topo
```

```python
new_ready_dates = {i: ready_dates[i] for i in range(1, n+1)}
for i in topo:
    if predecessors[i]:
        new_ready_dates[i] = max(ready_dates[i], max(new_ready_dates[j] + durations[j] for j in predecessors[i]))

new_ready_dates
```

```python
new_deadlines = {i: deadlines[i] for i in range(1, n+1)}

for i in reversed(topo):
    if successors[i]:
        new_deadlines[i] = min(new_deadlines[i], min(new_deadlines[s] - durations[s] for s in successors[i]))

new_deadlines
```

```python
from pysat.formula import WCNF, IDPool
from pysat.pb import PBEnc, EncType
from pysat.examples.rc2 import RC2

def solve_single_machine_with_predecessors(n, durations, ready_dates, deadlines, predecessors):

    # ----------------------------
    # Variables and time horizon
    # ----------------------------
    t_max = max(deadlines.values()) + 1
    jobs = list(range(1, n+1))

    vpool = IDPool()
    S = {}      # S[(i,t)] = var
    A = {}      # A[(i,t)] = var
    valid = {}  # valid start times per job

    wcnf = WCNF()

    print("\n=== VARIABLE GENERATION ===")

    for i in jobs:
        r_i = ready_dates[i]
        p_i = durations[i]
        dl_i = deadlines[i]

        last_start = dl_i - p_i
        if last_start < r_i:
            print(f"Job {i} is impossible!")
            return None, None

        valid[i] = list(range(r_i, last_start + 1))

        for t in valid[i]:
            S[(i, t)] = vpool.id(("S", i, t))

        for t in range(r_i, dl_i):
            A[(i, t)] = vpool.id(("A", i, t))

    print(f"Total variables: {vpool.top}")

    # ============================================
    # 1. Start Exactly Once
    # ============================================
    print("\n=== START EXACTLY ONCE ===")

    for i in jobs:
        lits = [S[(i, t)] for t in valid[i]]

        enc = PBEnc.equals(
            lits=lits,
            weights=[1]*len(lits),
            bound=1,
            encoding=EncType.binmerge
        )
        for cls in enc.clauses:
            wcnf.append(cls)

    # ============================================
    # 2. Link S -> A (activation constraints)
    # ============================================
    print("\n=== ACTIVATION CONSTRAINTS ===")

    for i in jobs:
        p_i = durations[i]
        for t_start in valid[i]:
            lit_s = S[(i, t_start)]
            for t in range(t_start, t_start + p_i):
                if t < t_max:
                    lit_a = A[(i, t)]
                    wcnf.append([-lit_s, lit_a])

    # ============================================
    # 3. Capacity: At most one job running at time t
    # ============================================
    print("\n=== MACHINE CAPACITY ===")

    for t in range(t_max):
        active = [A[(i, t)] for i in jobs if (i, t) in A]

        if len(active) > 1:
            enc = PBEnc.atmost(
                lits=active,
                weights=[1]*len(active),
                bound=1,
                encoding=EncType.binmerge
            )
            for cls in enc.clauses:
                wcnf.append(cls)

    # ============================================
    # 4. Precedence constraints (staircase)
    # predecessors[j] = list of jobs that must finish before j starts
    # ============================================
    print("\n=== PRECEDENCE CONSTRAINTS ===")

    for j in jobs:
        for i in predecessors.get(j, []):    # i → j
            p_i = durations[i]

            for t_j in valid[j]:
                for t_i in valid[i]:
                    if t_i + p_i > t_j:      # forbidden pair
                        wcnf.append([-S[(i, t_i)], -S[(j, t_j)]])

    # ============================================
    # SOLVE
    # ============================================
    print("\n=== SOLVING (HARD CONSTRAINTS ONLY) ===")

    solver = RC2(wcnf)
    model = solver.compute()

    if model is None:
        print("UNSAT — no feasible schedule.")
        return False, None

    # Extract chosen start times
    schedule = {}
    for (i, t), var in S.items():
        if model[var - 1] > 0:
            schedule[i] = t

    print("\nFeasible schedule found.")
    return True, schedule


solve_single_machine_with_predecessors(n, durations, new_ready_dates, new_deadlines, predecessors)
```

```python
def check_schedule(schedule, durations, ready_dates, deadlines, successors):
    ok = True
    print("\n=== Checking schedule ===")

    # --------------------------
    # 1. Ready date check
    # --------------------------
    print("\nReady date violations:")
    ready_errors = False
    for j, start in schedule.items():
        if start < ready_dates[j]:
            print(f"  Job {j}: start {start} < ready {ready_dates[j]}")
            ready_errors = True
            ok = False
    if not ready_errors:
        print("  ✓ None")

    # --------------------------
    # 2. Deadline check
    # --------------------------
    print("\nDeadline violations:")
    deadline_errors = False
    for j, start in schedule.items():
        finish = start + durations[j]
        if finish > deadlines[j]:
            print(f"  Job {j}: finish {finish} > deadline {deadlines[j]}")
            deadline_errors = True
            ok = False
    if not deadline_errors:
        print("  ✓ None")

    # --------------------------
    # 3. Precedence constraints
    # --------------------------
    print("\nPrecedence violations:")
    prec_errors = False
    for i in successors:
        for j in successors[i]:    # i → j
            if i in schedule and j in schedule:
                finish_i = schedule[i] + durations[i]
                if schedule[j] < finish_i:
                    print(f"  {i} → {j} violated: "
                          f"start[j]={schedule[j]} < finish[i]={finish_i}")
                    prec_errors = True
                    ok = False
    if not prec_errors:
        print("  ✓ None")

    print("\n=== Result:", "VALID ✔" if ok else "INVALID ✘", "===")
    return ok

check_schedule(schedule, durations, ready_dates, deadlines, successors)
```
